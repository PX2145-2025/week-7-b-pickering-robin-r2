```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import random, seed
```

```python
# Exercise 1
```

```python
# Part 1
```

```python
def walk(steps):
    '''
    steps is the length of the 1-D random walk vector returned
    Vectorised method.
    '''
    x = np.random.choice([-1, 1], size=steps) # random -1 or 1
    x = np.cumsum(x) # cumulative sum
    return x
```

```python
plt.plot(walk(1000) )
plt.ylabel("Position (x)")
plt.xlabel("Step number (i)")
plt.title("Position vs Step number")
```

```python
# Part 2
```

```python
for a in range(100):
  plt.plot(walk(1000) )

plt.ylabel("Position (x)")
plt.xlabel("Step number (i)")
plt.title("Position vs Step number")
```

```python
# Part 1
```

```python
steps = 1000
N = 10
particles = np.zeros((steps, N), int)

for a in range(steps):
    particles[a] = particles[a - 1] + np.random.choice([-1, 1], size=N)

for a in range(N):
    plt.plot(particles[:, a])

plt.title("Position vs Step number")
plt.xlabel("Step number (i)")
plt.ylabel("Position (x)")
```

```python
# Part 2
```

```python
steps = 1000
N = 500
particles = np.zeros((steps, N), int)

for a in range(1, steps):
    particles[a] = particles[a - 1] + np.random.choice([-1, 1], size=N)

for a in range(N):
    plt.plot(particles[:, a])

plt.title("Position vs Step number")
plt.xlabel("Step number (i)")
plt.ylabel("Position (x)")
```

```python
mean = np.mean(particles, axis=1)
mean_squared = np.mean(particles**2, axis=1)
print(mean[-1])
print(mean_squared[-1])
```

```python
# Part 3
```

```python
plt.plot(mean)
plt.title("Mean position vs Step number")
plt.xlabel("Step number (i)")
plt.ylabel("Mean position (x)")
```

```python
# Part 4
```

```python
plt.plot(mean_squared)
plt.title("Mean squared position vs Step number")
plt.xlabel("Step number (i)")
plt.ylabel("Mean squared position (x)")
```
